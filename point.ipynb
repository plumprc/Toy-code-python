```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def power_chain(A, x, t):
    points = []
    for i in range(t):
        points.append([x[0], x[1]])
        x = A.dot(x)
    return points
```

```python
A = np.array([[0.8,0],[0,0.64]])

# 设置画布
fig = plt.figure()
plt.xlim(-0.25, 0.25)
plt.ylim(-0.35, 0.35)

points = power_chain(A, np.array([0.2, 0.3]), 25)
for p in points:
    # plt.plot(p.x, p.y, "o")
    plt.plot(p[0], p[1])
    
# 画箭头
for i in range(len(points) - 1):
    dx = points[i+1][0] - points[i][0]
    dy = points[i+1][1] - points[i][1]
    plt.quiver(points[i][0], points[i][1], dx, dy, angles='xy', scale=1.03, scale_units='xy', width=0.006, color='r')
    
points = power_chain(A, np.array([-0.2, 0.3]), 25)
for p in points:
    # plt.plot(p.x, p.y, "o")
    plt.plot(p[0], p[1])
    
# 画箭头
for i in range(len(points) - 1):
    dx = points[i+1][0] - points[i][0]
    dy = points[i+1][1] - points[i][1]
    plt.quiver(points[i][0], points[i][1], dx, dy, angles='xy', scale=1.03, scale_units='xy', width=0.006, color='g')
    
points = power_chain(A, np.array([-0.2, -0.3]), 25)
for p in points:
    # plt.plot(p.x, p.y, "o")
    plt.plot(p[0], p[1])
    
# 画箭头
for i in range(len(points) - 1):
    dx = points[i+1][0] - points[i][0]
    dy = points[i+1][1] - points[i][1]
    plt.quiver(points[i][0], points[i][1], dx, dy, angles='xy', scale=1.03, scale_units='xy', width=0.006, color='b')
    
points = power_chain(A, np.array([0.2, -0.3]), 25)
for p in points:
    # plt.plot(p.x, p.y, "o")
    plt.plot(p[0], p[1])
    
# 画箭头
for i in range(len(points) - 1):
    dx = points[i+1][0] - points[i][0]
    dy = points[i+1][1] - points[i][1]
    plt.quiver(points[i][0], points[i][1], dx, dy, angles='xy', scale=1.03, scale_units='xy', width=0.006, color='orange')
    
plt.xlabel('Attractor')
plt.grid()
plt.savefig('track1.png')
plt.show()
```
